```python
import pandas as pd
```

```python
data_lahan = pd.read_csv('luaslahansawah.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_lahan.head()
```

Data lahan ini adalah data luas lahan pada tahun 2012 untuk setiap provinsi yang ada di Indonesia. Pada visualisasi data dibawah, ketahui bahwa tiga provinsi dengan luas lahan terbesar adalah Provinsi Jawa Timur, Jawa Tengah, dan Jawa Barat.

```python
data_lahan.columns
```

```python
tot = data_lahan.Luas.sum()
min = data_lahan.Luas.min()
max = data_lahan.Luas.max()

print(tot,min,max)
```

```python
me = data_lahan.mean()
md = data_lahan.median()

print(me)
```

```python
ax = data_lahan.groupby('Provinsi').mean()['Luas']
ax = ax.plot(kind='bar', title ="Luas Lahan Berdasarkan Nama Provinsi", figsize = (7,7), fontsize=9) #kalau pakai legend = True
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Luas(m)", fontsize=12)
```

```python
data_kemiskinan = pd.read_csv('processed-jumlah-lapangan-kerja-kepala-rumah-tangga-30persen-termiskin-indonesia-perjuli2012.csv', delimiter = ',', encoding = "ISO-8859-1")
```

```python
data_kemiskinan.columns
```

```python
data_kemiskinan[['nama_provinsi','lapangan_kerja']].head()
```

Data kemiskinan ini adalah data Rumah Tangga Miskin pada setiap provinsi dan setiap sektor lapangan kerja. Pada visualisasi data dibawah, diketahui bahwa RTM terbesar jumlahnya ada pada sektor Pertanian Padi dan Palawija, sedangkan RTM tertinggi terdapat pada Provinsi Jawa Timur, Jawa Tengah, dan Jawa Barat.

```python
ax = data_kemiskinan.groupby('lapangan_kerja').mean()['jumlah_kepala_rumah_tangga']
ax = ax.plot(kind='pie', title ="Jumlah Kepala Rumah Tangga Miskin Berdasarkan Lapangan Pekerjaan", figsize = (7,7), fontsize=9) #kalau pakai legend = True
```

```python
data_kemiskinan[data_kemiskinan['lapangan_kerja']=='Pertanian Tanaman Padi Dan Palawija'].head()
```

```python
total = data_kemiskinan.jumlah_kepala_rumah_tangga.sum()
mi = data_kemiskinan.jumlah_kepala_rumah_tangga.min()
ma = data_kemiskinan.jumlah_kepala_rumah_tangga.max()

print(total,mi,ma)
```

```python
me = data_kemiskinan.mean()
md = data_kemiskinan.median()

print(me)
```

```python
ax = data_kemiskinan.groupby('nama_provinsi').mean()['jumlah_kepala_rumah_tangga']
ax = ax.plot(kind='bar', title ="Jumlah Kepala Rumah Tangga Miskin Berdasarkan Nama Provinsi", figsize = (7,7), fontsize=9) #kalau pakai legend = True
ax.set_xlabel("Provinsi", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

Ada dua data yang digunakan dalam analisis ini. Yaitu data rumah tangga miskin dan data luas lahan. Didapati bahwa tiga provinsi dengan Rumah Tangga Miskin (RTM) terbanyak, pada tahun 2012, adalah Provinsi Jawa Barat, Jawa Tengah, dan Jawa Timur. Sedangkan dari visualisasi data yang menggambarkan lapangan kerja RTM, diketahui bahwa RTM dengan lapangan kerja pada sektor pertanian padi dan palawija ialah penyumbang terbesar RTM. Sektor pertanian padi dan palawija sendiri harus didukung oleh lahan sawah. Pada data luas lahan, menyatakan bahwa luas lahan di Provinsi Jawa Barat, Jawa Tengah, dan Jawa Timur, menjadi terbesar dibandingkan dengan provinsi lain.

Hal ini dapat menjadi indikasi awal bahwa jumah rumah tangga miskin tinggi pada sektor pertanian padi dan palawija, sekalipun lahan sawah garapan tergolong luas. Selanjutnya, dapat dimungkinkan bahwa banyaknya rumah tangga miskin yang bergantung pada sektor ini ada keterkaitan dengan kejenuhan sumber daya manusia. Hal yang dapat dilakukan untuk meningkatkan pendapatan rumah tangga, dengan melihat data ini, ialah dengan mempertahankan luas lahan sawah, namun mengalihkan lapangan kerja RTM dari sektor pertanian ke sektor lain.
