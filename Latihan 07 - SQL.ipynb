```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

## 1. Report the account representative for each customer.

```python
run("""
SELECT 
    CustomerName, 
    salesrepemployeenumber as SalesNumber, 
    Firstname as AccountRepFirstName,
    Lastname as AccountRepLastName
FROM
    customers t1
        INNER JOIN
    employees t2 ON t1.salesrepemployeenumber = t2.employeenumber
""")
```

## 2. Report total payments for Atelier graphique

```python
run("""
SELECT 
    CustomerNumber,
    CustomerName,
    sum(Amount) as TotalPayments
FROM
    customers
        INNER JOIN
    payments using (customernumber)
where
    customernumber = 103
""")
```

## 3. Report the total payments by date

```python
run("""
SELECT 
    paymentdate,
    SUM(amount) AS TotalBayar 
FROM payments 
GROUP BY paymentdate
 ORDER BY paymentdate""")
```

## 4. Report the products that have not been sold

```python
run("""
SELECT 
    t1.productCode, 
    productName, 
    ordernumber
FROM
    products t1
        left JOIN
    orderdetails t2 ON t1.productcode = t2.productcode
where ordernumber is null
""")
```

## 6. How many orders have been placed by Herkku Gifts?

## 7. Who are the employees in Boston?

```python
run("""
SELECT 
    firstname,
    lastname,
    c.officecode,
    city
FROM
    employees c
        INNER JOIN
    offices o ON c.officecode = o.officecode
Where
    city = 'boston'
""")
```

## 8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.

```python
run("""
SELECT 
    customername,
    c.customernumber,
    amount
FROM
    customers c
        INNER JOIN
    payments o ON c.customernumber = o.customernumber
Where
    amount > 100000
order by amount desc
""")
```

## 9. List the value of 'On Hold' orders.

```python
run("""
SELECT 
    o.ordernumber,
    quantityordered,
    priceeach,
    quantityordered*priceeach as value,
    status
FROM
    orderdetails o
        inner JOIN
    orders r ON o.ordernumber = r.ordernumber
having status = 'on hold'
order by value
""")
```

## 10.Report the number of orders 'On Hold' for each customer

```python
run("""
SELECT 
    c.customernumber,
    customername,
    r.ordernumber,
    count(quantityordered) as NumberofOnHoldOrders,
    status
FROM
    CUSTOMERS c
        INNER JOIN
    orders r ON c.customernumber = r.customernumber
        inner join
    orderdetails b on r.ordernumber = b.ordernumber
group by ordernumber
having status = 'on hold'
""")
```

## 1. List products sold by order date.

```python
run("""
SELECT 
    c.ordernumber,
    c.productcode,
    productname,
    orderdate
FROM
    orderdetails c
        INNER JOIN
    orders r ON c.ordernumber = r.ordernumber
        INNER JOIN
    products p on c.productcode = p.productcode
order by orderdate
""")
```

## 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck.

```python
run("""
SELECT 
    c.ordernumber,
    c.productcode,
    productname,
    orderdate
FROM
    orderdetails c
        INNER JOIN
    orders r ON c.ordernumber = r.ordernumber
        INNER JOIN
    products p on c.productcode = p.productcode
where productname = '1940 Ford Pickup Truck'
order by orderdate desc
""")
```

## 3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?

```python
run("""
SELECT 
    customername,
    s.ordernumber,
    sum(quantityordered*priceeach) as value
FROM
    customers c
        inner join
    orders s on c.customernumber = s.customernumber
        inner join
    orderdetails o on o.ordernumber = s.ordernumber
group by customername,ordernumber
having value >25000
order by value
""")
```

## 5. List the names of products sold at less than 80% of the MSRP

```python
run("""
SELECT 
    productCode, 
    productName, 
    buyprice,
    msrp
FROM
    products a
        LEFT JOIN
    ORDERDETAILS b using (productcode)
where buyprice < 80/100*msrp and ordernumber is not null
group by productcode
""")
```

## 6. Reports those products that have been sold with a markup of 100% or more (i.e., the priceEach is at least twice the buyPrice)

```python
run("""
SELECT 
    productCode, 
    productName, 
    buyprice,
    priceeach
FROM
    products a
        inner JOIN
    ORDERDETAILS b using (productcode)
where priceeach > 2*buyprice
order by buyprice
""")
```

## 7. List the products ordered on a Monday.

```python
run("""
SELECT 
    productname,
    a.ordernumber,
    a.productcode,
    orderdate,
    dayname(orderdate) as Day
FROM
    orderdetails a
        inner JOIN
    orders b on a.ordernumber = b.ordernumber
        inner join
    products c on a.productcode = c.productcode
where dayname(orderdate) = 'monday'
""")
```
