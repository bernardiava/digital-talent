```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
run("SHOW TABLES")
```

## 1. Prepare a list of offices sorted by country, state, city.

```python
run("""
SELECT officeCode,country,state,city 
FROM offices
ORDER BY country , state, city
""")
```

## 5. Report total payments for October 28, 2004

```python
run("""
SELECT 
    YEAR(paymentDate) AS Tahun,
    MONTH(paymentDate) AS Bulan,
    DAY(paymentDate) AS Hari,
    SUM(amount) AS TotalBayar 
FROM payments 
WHERE DAY(paymentDate)=28 AND MONTH(PAYMENTDATE)=9 AND YEAR(PAYMENTDATE)= 2004
GROUP BY Tahun,Bulan,Hari
 ORDER BY BULAN,HARI""")
```

## 6. Report those payments greater than \$100,000

```python
run("""
SELECT 
    YEAR(paymentDate) AS Tahun,
    MONTH(paymentDate) AS Bulan,
    SUM(amount) AS TotalBayar 
FROM payments 
GROUP BY Tahun,Bulan
HAVING TotalBayar > 100000
ORDER BY TotalBayar""")
```

## 13. Report the name and city of customers who don't have sales representatives?

```python
run("""
SELECT 
    customerName, 
    city
FROM
    customers
WHERE
    city NOT LIKE '%Francisco%' AND city NOT LIKE '%boston%' AND city NOT LIKE '%nyc%' 
    AND city NOT LIKE '%paris%' AND city NOT LIKE '%tokyo%' AND city NOT LIKE '%sydney%'
    AND city NOT LIKE '%london%'
""")
```

## 15. Which orders have a value greater than $5,000?

```python
run("""
SELECT 
    quantityordered AS quantity,
    priceeach AS priceeachproduct,
    quantityordered*priceeach AS TotalBayar 
FROM orderdetails
where quantityordered*priceeach > 5000
ORDER BY TotalBayar
""")
```
